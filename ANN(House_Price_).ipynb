```python
import pandas as pd
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('/content/housepricedata.csv')
df.head()
```

```python
x = df.iloc[:,0:10]
y=df.iloc[0:,10]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scale = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scale,y,test_size=0.3,random_state=1)
```

```python
model = Sequential()
# model.add(Dense(32,input_dim=10,activation='relu'))
```

```python
input_layer = Dense(20,input_shape=(10,))
model.add(input_layer)
```

```python
hidden_layer1 = Dense(20,activation='relu')
model.add(hidden_layer1)
```

```python
hidden_layer2 = Dense(20,activation='relu')
model.add(hidden_layer2)
```

```python
out_layer = Dense(1,activation='sigmoid')
model.add(out_layer)
```

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=10)
```

```python
val_los,val_acc = model.evaluate(x_test,y_test)
print(val_los,val_acc)
```

```python
lst = [[14260,8,5,1145,2,1,4,9,1,836]]
```

```python
scaled_data = scaler.transform(lst)
```

```python
result = model.predict(scaled_data)
```

```python
if(result>0.5):
  print('House is expensive')
else:
  print('House is cheap')
```
