```python
#We are given data like age and what insurance apply logistic regression model and predict whether a person takes insurance or not based on his age.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('insurance_data.csv')
data.head()
```

```python
x = data.iloc[:,:-1].values 
```

```python
y = data.iloc[:,1].values  
```

```python
plt.xlabel('Age')
plt.ylabel('Have Insurence')
plt.scatter(x,y,marker = '+' , color = 'red')
```

```python
sns.regplot(x = 'age',y = 'bought_insurance',data = data , logistic = True , marker = '+' , color = 'red')
```

```python
model = LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[20]])
```

```python
model.predict([[40]])
```
