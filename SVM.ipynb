```python
#Find the iris flowers data set using SVM and find out the flower type, depending on the given input data.
# 0 --> Setosa , 1 --> Versicolor, 2 --> Virginica
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.data
iris.target
iris.target_names
iris.feature_names
```

```python
import pandas as pd 
```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
data.head()
```

```python
data['target'] = iris.target
data.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
sns.pairplot(data,hue='target',palette="brg")
plt.show()
```

```python
x = data.drop(['target'],axis='columns')
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model.predict([iris.data[50]])  #--> versicolor
```

```python
model.predict([[7.7,2.6,6.9,2.3]]) #--> Virginica
```

```python
model = SVC(kernel='linear')
model.fit(x_train, y_train)
model.score(x_test,y_test)
```
